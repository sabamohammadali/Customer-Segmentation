We will scale all the features and do one-hot-encoding for gender

```python
ss = StandardScaler()
```

```python
df_t = df.copy()

ss = StandardScaler()
df_t['Age'] = ss.fit_transform(df['Age'].values.reshape(-1,1))
df_t['Income'] = ss.fit_transform(df['Income'].values.reshape(-1,1))
df_t['Score'] = ss.fit_transform(df['Score'].values.reshape(-1,1))
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
sns.scatterplot(x=df_t.Age, y=df_t.Income, hue=df_t.Gender)
plt.title('Age vs Income - Scaled')

plt.subplot(1,3,2)
sns.scatterplot(x=df_t.Age, y=df_t.Score, hue=df_t.Gender)
plt.title('Age vs Score - Scaled')

plt.subplot(1,3,3)
sns.scatterplot(x=df_t.Income, y=df_t.Score, hue=df_t.Gender)
plt.title('Income vs Score - Scaled')

plt.show()
```

There is no difference between scaled and original data points

```python
df = pd.get_dummies(df, columns=['Gender'])
```

We will plot the inertia to find the best number of clusters

```python
inertia = []
for i in range(1, 10):
    km = KMeans(n_clusters=i)
    km.fit(df)
    inertia.append(km.inertia_)
```

```python
plt.figure(figsize=(20,5))
sns.lineplot(x=range(1,10), y=inertia)
plt.title('KMeans inertia')
plt.show()
```

2, 3, 5 and 6 seem to be the best number of clusters. We will try kmeans with 2, 3, 4, 5 and 6 number of clusters and analyse the results

```python
clusters = [2,3,4,5,6]
cluster_centers = {}
for c in clusters:
    km = KMeans(n_clusters=c).fit(df)
    df['cluster' + str(c)] = km.labels_
    cluster_centers[str(c)] = km.cluster_centers_
```

```python
df.head()
```

```python
plt.figure(figsize=(20,15))
for i, c in enumerate(clusters):
    plt.subplot(2,3,i+1)
    sns.scatterplot(df.Income, df.Score, df['cluster' + str(c)], s=120, palette=sns.color_palette("tab10", c))
    sns.scatterplot(cluster_centers[str(c)][:,1], cluster_centers[str(c)][:,2], color='red', s=300)
    plt.title('Number of clusters: ' + str(c))
    
plt.show()
```

```python
plt.figure(figsize=(20,15))
for i, c in enumerate(clusters):
    plt.subplot(2,3,i+1)
    sns.scatterplot(df.Income, df.Age, df['cluster' + str(c)], s=120, palette=sns.color_palette("tab10", c))
    sns.scatterplot(cluster_centers[str(c)][:,1], cluster_centers[str(c)][:,2], color='red', s=300)
    plt.title('Number of clusters: ' + str(c))
    
plt.show()
```

```python
plt.figure(figsize=(20,15))
for i, c in enumerate(clusters):
    plt.subplot(2,3,i+1)
    sns.scatterplot(df.Age, df.Score, df['cluster' + str(c)], s=120, palette=sns.color_palette("tab10", c))
    sns.scatterplot(cluster_centers[str(c)][:,1], cluster_centers[str(c)][:,2], color='red', s=300)
    plt.title('Number of clusters: ' + str(c))
    
plt.show()
```

It is evident that Income and Spending Score have the more prominent effect of the customer segmentation. The ideal number of clusters is 5

```python
plt.figure(figsize=(8,8))
sns.scatterplot(df.Income, df.Score, df['cluster5'], s=120, palette=sns.color_palette("tab10", 5))
plt.title('Income vs Score')

plt.show()
```

These clusters can be understood as following:

Cluster 0: People with low spending but high score so they love to spend (Can be targetted by the mall as they seem to be easily influenced)

Cluster 1: People with high income but low spending score (Do not like to spend so probably won't be influenced by sales and schemes)

Cluster 2: People with mediocre income and mediocre spending score

Cluster 3: People with high income and high spending score (Best people to be targetted by the mall for schemes and discounts

Cluster 4: People with low income and low spending (Not ideal to be targetted by the mall)
